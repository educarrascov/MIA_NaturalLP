Definimos reglas básicas de chunking en Español para una gramática parcial simple. Asumimos el uso de etiquetas POS que provee un texto etiquetado con SpACY:

**Frase Nominal** (**FN**) se define como DET NOUN* ADJ*  
....

Luego, la especificación es la siguiente:

```python
gramatica = '''                                                                                                              
             FN:                                                                                                                    
                 {<DET>(<NOUN>)*<ADJ>*}
             NVN:                                                                                                                  
             {<np000000><VB.*><NP>}                                                                                                    
            '''
```

Re-usamos nuestra función de POS tagging definida en otras secciones para entregar pares (palabra,POS) para cada palabra de un tetxo:

```python
def Etiquetar(texto):
   doc = nlp(texto)
   Etiquetado = [(t.text,t.pos_) for t in doc]
   return(Etiquetado)
```

Luego, definimos la función **Chunking(texto,gramatica)**, que nos realiza análisis parcial de un texto desde reglas de chunking, retornando los árboles que calzan:

```python
def Chunking(texto,gramatica):
  tagged = Etiquetar(texto)    
  chunker = nltk.chunk.RegexpParser(gramatica)
  Arbol   = chunker.parse(tagged)
  for subarbol in Arbol.subtrees():
        # Determinar si hay alguna parte del texto que calce con la regla FN
        if subarbol.label() == 'FN': 
            print(subarbol)
```

Ahora, invocamos nuestro programa principal, ajustando la ruta (PATH) donde se encuentra el texto de prueba:

```python
PATH = 'sample.txt'
nlp = es_core_news_sm.load()
texto = open(PATH, encoding="utf8").read()
Chunking(texto,gramatica)
```
